# **Importing  the required modules**

```python
from PIL import Image
import matplotlib.pyplot as plt
```

# **Loading the image**

```python
# load the image 
img = Image.open('image.png')
# display the image
plt.imshow(img)
# also tell the size of the image
print(img.size)
```

```python
# now we will apply the filters for edge detection
from PIL import ImageFilter
# apply the filter
img_edge = img.filter(ImageFilter.FIND_EDGES)
# display the image
plt.imshow(img_edge)
```

```python
# now we will apply convolutional filter
from PIL import ImageFilter
# apply the filter
img_conv = img.filter(ImageFilter.CONTOUR)
# display the image
plt.imshow(img_conv)
```

```python
# adding padding to the image
from PIL import ImageOps
# add padding to the image
img_pad = ImageOps.expand(img, border=20, fill='black')
# display the image
plt.imshow(img_pad)
```

```python
# max pooling the image
from PIL import ImageFilter
# apply the filter
img_pool = img.filter(ImageFilter.MaxFilter(3))
# display the image
plt.imshow(img_pool)
```
