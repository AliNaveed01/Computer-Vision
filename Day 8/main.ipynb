# **Importing Required Modules**

```python
import cv2
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
# load the image check.jpg from the current directory
image = cv2.imread('check.jpg')
# convert the image from BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
# now take a small part of the picture and show it
# the part is the top left corner of the picture
small_part = image[300:450, 600:700]
plt.imshow(small_part)

small_part_flip_horizontal = cv2.flip(small_part, 1)
plt.imshow(small_part_flip_horizontal)
```

```python
# now we will create a function which we will use to overlapn the subnet on the image at different locations and print them
def show_image_with_subnet(image, subnet, x, y):
    image_copy = image.copy()
    image_copy[y:y+subnet.shape[0], x:x+subnet.shape[1]] = subnet
    plt.imshow(image_copy)
    plt.show()
```

```python
# printing the image with the subnet at different locations
show_image_with_subnet(image, small_part, 100, 100)
```

```python
show_image_with_subnet(image, small_part, 930, 280)
show_image_with_subnet(image, small_part_flip_horizontal, 750, 240)

```

```python
def blend_images(image, small_part_flip_horizontal, position):
    image_copy = image.copy()
    y, x = position
    # now we will subnet the image by averaging the pixel values of the two images
    image_copy[y:y+small_part_flip_horizontal.shape[0], x:x+small_part_flip_horizontal.shape[1]] = (image_copy[y:y+small_part_flip_horizontal.shape[0], x:x+small_part_flip_horizontal.shape[1]] + small_part_flip_horizontal) / 2
    plt.imshow(image_copy)

# now we will blend the images at different positions
blend_images(image, small_part_flip_horizontal, (100, 100))
```
